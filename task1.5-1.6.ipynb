```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math

sns.set_theme(style="darkgrid")
```

```python
def F1(n: int) -> float:
    return 0


def Hn(A: float, R: float, delta: float, H_prev: float, F_prev: float) -> float:
    return (1 - delta / (A * R)) * H_prev + delta / A * F_prev
```

```python
def do_task_a():
    
    n = 10
    A = 5
    R = 1
    delta = 2.5
    
    H1 = np.zeros(n, dtype=float)
    H1[0] = 10
    
    for i in range(0, n - 1, 1):
        H1[i + 1] = Hn(A, R, delta, H1[i], F1(i))
        
    return H1
```

```python
def do_task_b():
    
    n = 25
    A = 5
    R = 1
    delta = 1
    
    H2 = np.zeros(n, dtype=float)
    H2[0] = 10
    
    for i in range(0, n - 1, 1):
        H2[i + 1] = Hn(A, R, delta, H2[i], F1(i))

    return H2
```

```python
def do_task_c():
    
    n = 100
    A = 5
    R = 1
    delta = 0.25
    
    H3 = np.zeros(n, dtype=float)
    H3[0] = 10
    
    for i in range(0, n - 1, 1):
        H3[i + 1] = Hn(A, R, delta, H3[i], F1(i))
        
    return H3
```

```python
H1 = do_task_a()
H2 = do_task_b()
H3 = do_task_c()
```

```python
sns.lineplot(data=H1)
plt.show()
```

```python
sns.lineplot(data=H2)
plt.show()
```

```python
sns.lineplot(data=H3)
plt.show()
```

```python
def do_task_d():
    n = min(H1.shape[0], min(H2.shape[0], H3.shape[0]))

    sns.lineplot(data=H1)
    sns.lineplot(data=H2)
    sns.lineplot(data=H3)
    
    plt.xlim([0, n - 1])
    plt.show()

do_task_d()
```

# Task 1.6

$F_0=c*H^{\frac{1}{2}}, c = 3$ 

$H_A(t + \Delta t) = H_A(t)*(1 + \frac{\Delta t * c}{A} * H_A^{-\frac{1}{2}}(t)) + \frac{\Delta t}{A}*F_1(t)$

```python
c = 3

def F0(H_prev: float) -> float:
    return c*H_prev**1/2

def Hn_mod(A: float, delta: float, H_prev: float, F_prev: float, F0_prev: float) -> float:
    return H_prev + 1/A * (F_prev - F0_prev)*delta
```

```python
def do_task_a_mod():
    
    n = 10
    A = 5
    R = 1
    delta = 2.5
    
    H1_mod = np.zeros(n, dtype=float)
    H1_mod[0] = 10
    
    for i in range(0, n - 1, 1):
        H1_mod[i + 1] = Hn_mod(A,delta, H1_mod[i], F1(i), F0(H1_mod[i]))
        
    return H1_mod
```

```python
def do_task_b_mod():
    
    n = 25
    A = 5
    R = 1
    delta = 1
    
    H2_mod = np.zeros(n, dtype=float)
    H2_mod[0] = 10
    
    for i in range(0, n - 1, 1):
        H2_mod[i + 1] = Hn_mod(A, delta, H2_mod[i], F1(i), F0(H2_mod[i]))

    return H2_mod
```

```python
def do_task_c_mod():
    
    n = 100
    A = 5
    R = 1
    delta = 0.25
    
    H3_mod = np.zeros(n, dtype=float)
    H3_mod[0] = 10
    
    for i in range(0, n - 1, 1):
        H3_mod[i + 1] = Hn_mod(A, delta, H3_mod[i], F1(i), F0(H3_mod[i]))
        
    return H3_mod
```

```python
H1_mod = do_task_a_mod()
H2_mod = do_task_b_mod()
H3_mod = do_task_c_mod()
```

```python
sns.lineplot(data=H1)
sns.lineplot(data=H1_mod)
plt.show()
```

```python
sns.lineplot(data=H2)
sns.lineplot(data=H2_mod)
plt.show()
```

```python
sns.lineplot(data=H3)
sns.lineplot(data=H3_mod)
plt.show()
```

```python
def do_task_d_mod():
    n = min(H1_mod.shape[0], min(H2_mod.shape[0], H3_mod.shape[0]))

    sns.lineplot(data=H1_mod)
    sns.lineplot(data=H2_mod)
    sns.lineplot(data=H3_mod)
    
    plt.xlim([0, n - 1])
    plt.show()
    
do_task_d_mod()
```

```python
def compare():
    
    sns.lineplot(data=H1)
    sns.lineplot(data=H2)
    sns.lineplot(data=H3)
    
    sns.lineplot(data=H1_mod)
    sns.lineplot(data=H2_mod)
    sns.lineplot(data=H3_mod)
    
    n = min(H1_mod.shape[0], min(H2_mod.shape[0], H3_mod.shape[0]))
    m = min(H1.shape[0], min(H2.shape[0], H3.shape[0]))
    
    plt.xlim([0, min(n, m) - 1])
    plt.show()
    
compare()
```
