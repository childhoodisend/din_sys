# Task 1.5

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
sns.set(rc={"figure.figsize":(15, 8)}) #width=15, #height=8
```

```python
delta_t = np.array([2.5, 1, 0.25], dtype=float)
nums    = np.array([10, 25, 100], dtype=int)
A = 5
R = 1

H1 = np.zeros(nums[0] + 1, dtype=float)
H2 = np.zeros(nums[1] + 1, dtype=float)
H3 = np.zeros(nums[2] + 1, dtype=float)

H1_mod = np.zeros(nums[0] + 1, dtype=float)
H2_mod = np.zeros(nums[1] + 1, dtype=float)
H3_mod = np.zeros(nums[2] + 1, dtype=float)

H1[0] = H2[0] = H3[0] = H1_mod[0] = H2_mod[0] = H3_mod[0] = 10


time_grid1 = np.array([delta_t[0] * i for i in range(nums[0] + 1)], dtype=float)  
time_grid2 = np.array([delta_t[1] * i for i in range(nums[1] + 1)], dtype=float)
time_grid3 = np.array([delta_t[2] * i for i in range(nums[2] + 1)], dtype=float)
```

$H_A(t + \Delta t) = (1 - \frac{\Delta t} {A * R}) * H_A(t) + \frac{\Delta t}{A} * F_1(t)$ 

$H_A(t + \Delta t) = (1 - \frac{\Delta t} {A * R}) * H_A(t)$ ัะบ $F_1(t) = 0$

```python
def Hn(delta: float, H_prev: float) -> float:
    return max(0, (1 - delta / (A * R)) * H_prev)
    
```

```python
def do_task_a():
    
    for i in range(0, len(time_grid1) - 1, 1):
        H1[i + 1] = Hn(delta_t[0], H1[i])
```

```python
def do_task_b():
    
    for i in range(0, len(time_grid2) - 1, 1):
        H2[i + 1] = Hn(delta_t[1], H2[i])
```

```python
def do_task_c():
    
    for i in range(0, len(time_grid3) - 1, 1):
        H3[i + 1] = Hn(delta_t[2], H3[i])
        
```

```python
do_task_a()
do_task_b()
do_task_c()
```

```python
sns.lineplot(x=time_grid1, y = H1)
plt.show()
```

```python
sns.lineplot(x=time_grid2, y = H2)
plt.show()
```

```python
sns.lineplot(x=time_grid3, y = H3)
plt.show()
```

```python
def do_task_d():
    
    sns.lineplot(x=time_grid1, y = H1)
    sns.lineplot(x=time_grid2, y = H2)
    sns.lineplot(x=time_grid3, y = H3)
    
    plt.legend(['n = 10', 'n = 25', 'n = 100'])
    plt.show()

do_task_d()
```

# Task 1.6

$F_0=c*H^{\frac{1}{2}}, c = 3$ 

$H_A(t + \Delta t) = H_A(t)*(1 - \frac{\Delta t * c}{A} * H_A^{-\frac{1}{2}}(t))$,

ัะบ $F_1(t) = 0$

```python
c = 3

def Hn_mod(delta: float, H_prev: float) -> float:
    return max(0, H_prev - delta * np.sqrt(H_prev)* c / A)
    
```

```python
def do_task_a_mod():
    
    for i in range(0, len(time_grid1) - 1, 1):
        H1_mod[i + 1] = Hn_mod(delta_t[0], H1_mod[i])
```

```python
def do_task_b_mod():
    
    for i in range(0, len(time_grid2) - 1, 1):
        H2_mod[i + 1] = Hn_mod(delta_t[1], H2_mod[i])
```

```python
def do_task_c_mod():
    
    for i in range(0, len(time_grid3) - 1, 1):
        H3_mod[i + 1] = Hn_mod(delta_t[2], H3_mod[i])
```

```python
do_task_a_mod()
do_task_b_mod()
do_task_c_mod()
```

```python
sns.lineplot(x=time_grid1, y = H1)
sns.lineplot(x=time_grid1, y = H1_mod)
plt.legend(['H1', 'H1_mod'])
plt.show()
```

```python
sns.lineplot(x=time_grid2, y = H2)
sns.lineplot(x=time_grid2, y = H2_mod)
plt.legend(['H2', 'H2_mod'])
plt.show()
```

```python
sns.lineplot(x=time_grid3, y = H3)
sns.lineplot(x=time_grid3, y = H3_mod)
plt.legend(['H3', 'H3_mod'])
plt.show()
```

```python
def do_task_d_mod():
    
    sns.lineplot(x=time_grid1, y = H1_mod)
    sns.lineplot(x=time_grid2, y = H2_mod)
    sns.lineplot(x=time_grid3, y = H3_mod)
    
    plt.legend(['n = 10', 'n = 25', 'n = 100'])
    
    plt.show()
    
do_task_d_mod()
```

```python
def compare():
    
    sns.lineplot(x=time_grid1, y = H1)
    sns.lineplot(x=time_grid2, y = H2)
    sns.lineplot(x=time_grid3, y = H3)
  
    sns.lineplot(x=time_grid1, y = H1_mod)
    sns.lineplot(x=time_grid2, y = H2_mod)
    sns.lineplot(x=time_grid3, y = H3_mod)
    
    plt.legend(['H1', 'H2', 'H3', 'H1_mod', 'H2_mod', 'H3_mod'])
    plt.show()
    
compare()
```
